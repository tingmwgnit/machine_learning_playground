```python
#load the data
train = pd.read_csv('train.csv')
test  = pd.read_csv('test.csv')
```

# 4. EDA

# 4.1 Distribution of sale price

```python
sns.displot(train['SalePrice'])
```

# 4.2 Top 3 factors in house price:
1. Overall material and finish quality
2. Above grade (ground) living area square feet
3. Size of garage in car capacity

```python
#correlation matrix
corrmat = train.corr()
```

```python
#top 3 factors of price
k = 4 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
# keep only the top 3 factors & drop any rows with missing values
key_train = train[['OverallQual','GrLivArea','GarageCars','SalePrice']]
key_test = test[['OverallQual','GrLivArea','GarageCars']]
```

# 5. Data Cleaning


```python
#drop rows with any missing values
key_train.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
key_train.isnull().sum().max()
```

```python
# the column 'GarageCars' has missing values
key_test.isnull().any()
```

```python
# found it! the NaN value is in the 1116th row
key_test[key_test['GarageCars'].isnull()]
```

```python
# will replace it with the median value of GarageCars since we cannot delete any rows in the test dataset othervise 
# it will mess the Kaggle competition submission up
key_test['GarageCars'].iloc[1116] = key_test['GarageCars'].median()
```

# 5. Feature Engineering

# 6. Model Fitting

```python
# create regressor object
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
 
# fit the regressor with x and y data
regressor.fit(key_train[list(key_train.columns[0:-1])], key_train['SalePrice'])
```

```python
result = regressor.predict(key_test)
```

```python
type(result)
```

# 7. Create Submission to Kaggle competition

```python
submission = pd.DataFrame(test['Id'], columns=['Id'])
submission['SalePrice'] = result
```

```python
submission
```

```python
submission.to_csv('submission.csv',index=False) #ranked 3812/4653   Score: 0.20666
```
